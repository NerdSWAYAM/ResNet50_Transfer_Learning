```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.preprocessing import image

# 1. Load model
model = keras.models.load_model("resnet50_transfer_best.h5")

# 2. Load and preprocess image
IMG_SIZE = (224, 224)  # change to match your model's input

img = image.load_img("images (1).jpeg", target_size=IMG_SIZE)
img_array = image.img_to_array(img)          # shape: (H, W, 3)
img_array = np.expand_dims(img_array, axis=0)  # shape: (1, H, W, 3)
img_array = img_array / 255.0                   # normalize to [0, 1]

# 3. Predict
predictions = model.predict(img_array)
predicted_class = np.argmax(predictions[0])
confidence = np.max(predictions[0]) * 100

print(f"Predicted class index: {predicted_class}")
print(f"Confidence: {confidence:.2f}%")
```

```python
import os

# Point this to your training data folder
data_dir = "./train"
CLASS_NAMES = sorted(os.listdir(data_dir))  # alphabetical = same order Keras used

predicted_label = CLASS_NAMES[predicted_class]
print(f"Predicted: {predicted_label}")
```
